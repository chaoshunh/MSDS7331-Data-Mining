## DESCRIPTION OF DATA

The City of Dallas has a strong commmitment to transparency and accountability. Citizens or interested parties are provided an invaluable resource to easily access data published by the City. This data acts as a bridge between the citizens of Dallas and the city. The information contained in this dataset allows Dallasites to access accurate information in a timely manner. Crime classifications are based on information supplied to the Dallas Police Department. + Sexually oriented offenses

+ Listing of property items that are considered evidence

+ Social Service Referral offenses

+ Identifying vehicle information in certain offenses

+ From June 1, 2014 to September 7, 2018

#### VICTIM RACE (Ethnicity)
+ Victim_Race was aligned with the US Census Bureau standards on race and ethnicity as below.  However, we did deviate from the US Census categories and retained Latino or Hispanic as a distinct ethnic group and did not merge it with American Indian or Alaska Native
    
    * **White** â€“ A person having origins in any of the original peoples of Europe, the Middle East, or North Africa.
        
    * **Black or African American** â€“ A person having origins in any of the Black racial groups of Africa.
        
    * **American Indian or Alaska Native** â€“ A person having origins in any of the original peoples of North and South America (including Central America) and who maintains tribal affiliation or community attachment.
        
    * **Asian** â€“ A person having origins in any of the original peoples of the Far East, Southeast Asia, or the Indian subcontinent including, for example, Cambodia, China, India, Japan, Korea, Malaysia, Pakistan, the Philippine Islands, Thailand, and Vietnam.
        
    * **Native Hawaiian or Other Pacific Islander** â€“ A person having origins in any of the original peoples of Hawaii, Guam, Samoa, or other Pacific Islands.
        
    * **Latino or Hispanic** â€“ A person from Mexico or South and Central America
        
*(Source: US Census Bureau)*

```python
#--------------------------------------------------------------------------------
# CLEANING AND STANDARDIZING THE VICTIM_RACE ATTRIBUTE
#--------------------------------------------------------------------------------
print('-------------------------------------------------')
print('Victim_Race BEFORE cleaning and standardization')
print('-------------------------------------------------')
print(incident['Victim_Race'].unique())

#----------------------------------------------------------------------------------
# Standardize as per US Census naming and replace missing and other with "Unknown"
#----------------------------------------------------------------------------------
incident.loc[incident['Victim_Race'] == 'American Indian / Native American', 'Victim_Race'] = 'American Indian or Alaska Native'
incident.loc[incident['Victim_Race'] == 'American Indian/Native American', 'Victim_Race'] = 'American Indian or Alaska Native'
incident.loc[incident['Victim_Race'] == 'Asian or Pacific Islander', 'Victim_Race'] = 'Native Hawaiian or Other Pacific Islander'
incident.loc[incident['Victim_Race'] == 'Native Hawaiian or Pacific Islander', 'Victim_Race'] = 'Native Hawaiian or Other Pacific Islander'
incident.loc[incident['Victim_Race'] == 'Latin / Hispanic', 'Victim_Race'] = 'Latino or Hispanic'
incident.loc[incident['Victim_Race'] == 'Latin/Hispanic', 'Victim_Race'] = 'Latino or Hispanic'
incident.loc[incident['Victim_Race'] == 'Black', 'Victim_Race'] = 'Black or African American'
incident.loc[incident['Victim_Race'] == 'TEST', 'Victim_Race'] = 'Unknown'
incident.loc[incident['Victim_Race'] == 'N', 'Victim_Race'] = 'Unknown'
incident['Victim_Race'].fillna('Unknown', inplace=True)

#----------------------------------------------------------------------------------
# Add counts and sort
#----------------------------------------------------------------------------------
print('------------------------------------------------')
print('Victim_Race AFTER cleaning and standardization')
print('------------------------------------------------')
incident['Victim_Race'].value_counts().sort_values()
```

#### CITY
+ Standardization of city names i.e. Dallas


```python
#-----------------------------------------------------------------
# Standardization of city names i.e. Dallas
#-----------------------------------------------------------------
incident.loc[:,'City'] = incident['City'].fillna("MISSING")

city_replace = dict.fromkeys(['Dallas','DALAS','DALLS','DALLLAS','DALLA','DLS',
                              'DALALS','D','DAL','DLLAS','DSLLAS','DALLAST','DAQLLAS'], 'DALLAS')

incident.loc[:,'City']= incident['City'].replace(city_replace)
```

### Add New Calculated Columns

```python
#-----------------------------------------------------------------
# Coding Binary variables for some columns  
#-----------------------------------------------------------------
def get_binary_code(row):
    if row == 'Yes':
        return 1
    else:
        return 0

#-----------------------------------------------------------------    
# Arrest_status variable binary coding
#-----------------------------------------------------------------
def get_status(row):
    if row['Offense_Status']=='Clear by Arrest':
        return 1
    elif row['Offense_Status']=='Clear by Exceptional Arrest':
        return 1
    else:
        return 0

#-----------------------------------------------------------------
# Converting string to datetimes
#-----------------------------------------------------------------

incident.loc[:,'Call_Received']= pd.to_datetime(incident['Call_Received_Date_Time'], 
                                                format="%m/%d/%Y %I:%M:%S %p")
incident.loc[:,'Call_Cleared']= pd.to_datetime(incident['Call_Cleared_Date_Time'], 
                                               format="%m/%d/%Y %I:%M:%S %p")
incident.loc[:,'Call_Dispatch']= pd.to_datetime(incident['Call_Dispatch_Date_Time'], 
                                                format="%m/%d/%Y %I:%M:%S %p")

#----------------------------------------------------------------------------------
# Parsing number of offenses per incident from 'Service_Number_ID' data definition
#----------------------------------------------------------------------------------
incident.loc[:,'Number_of_offense']= incident['Service_Number_ID'].str.split("-",
                                                                             expand=True).rename(columns = {2:'Number_of_offense'})


# incident.loc[:,'Year_of_offense']= incident['Service_Number_ID'].str.split("-",
#                                                                              expand=True).rename(columns = {1:'Year_of_offense'})

#----------------------------------------------------------------------------------
# Calculating response time from Call Received to Call Dispatch from original 
# columns 'Call_Received_Date_Time' and 'Call_Dispatch_Date_Time'
#----------------------------------------------------------------------------------
incident.loc[:,'Response_time'] = pd.Series(data=(incident['Call_Dispatch']-incident['Call_Received']).astype('timedelta64[m]').astype('float'),
                                          index=incident.index,name='Response_Time')


#-----------------------------------------------------------------
# Parsing Latitude and Longitude values from 'Location1 attribute 
#-----------------------------------------------------------------
incident.loc[:,'Latitude']= incident['Location1'].str.split("\n",
                                                            expand=True)[2].str.split(',',
                                                                                      expand=True)[0].str.split('(',
                                                                                                                expand=True).rename(columns = {1 : 'Latitude'}).astype(int, errors='ignore')
                                                                                                  
incident.loc[:,'Longitude']= incident['Location1'].str.split("\n",
                                                            expand=True)[2].str.split(',',
                                                                                      expand=True)[1].str.split(')',
                                                                                                                expand=True).rename(columns = {0 : 'Longitude'}).astype(int, errors='ignore')

#------------------------------------------------------------
# Incident arrest status based on "Offense_Status" attribute
#------------------------------------------------------------
incident.loc[:,'Arrest_status']= incident.apply(get_status, axis=1) 

#--------------------------------------------
# Applying Binary Coding to these attributes
#--------------------------------------------
incident.loc[:,'Hate_Crime'] = incident['Hate_Crime'].apply(get_binary_code)
incident.loc[:,'Gang_Related_Offense'] = incident['Gang_Related_Offense'].apply(get_binary_code)
incident.loc[:,'Drug_Related_Istevencident'] = incident['Drug_Related_Istevencident'].apply(get_binary_code)
```

- The following attributes were either parsed or calculated from other attributes
    + Number_of_Offense: Number of offenses per reported incident number
    + Response_Time: This is the time between receiving the call and dispatching it to officers for investigation
    + Call_Received
    + Call_Cleared
    + Call_Dispatch

### Outlier Identification and Cleanup
- The Response time had outliers that needed to be addressed

```python
#--------------------------------------------------
# Simple Statistics of the Response time attribute
#--------------------------------------------------
incident['Response_time'].describe()
```

**ASSUMPTION**: *It is assumed that all calls received are dispatched within a 24 hour window for officer investigation*

From the summary statistics results for the response attribute we observe the following:
- Average response time is approximately 37 minutes 
- It appears that the min is negative which does not make practical sense
- The max value also seems rather high at 13290 minutes which equates to 221 hours or 9 days
- Clearly these outliers need to be addressed before progressing with further analysis

```python
#------------------------------------------------------
# Plot IQR for Response Time - 25th and 75th quartiles
#------------------------------------------------------
q75, q25 = np.percentile(incident.Response_time.dropna(), [75 ,25])

#--------------------
# Calculate the IQR
#--------------------
iqr = q75 - q25

#-----------------------------------------------------
# Establish the limits for the outliers using the IQR
#-----------------------------------------------------
min = q25 - (iqr*1.5)
max = q75 + (iqr*1.5)

print('Minimum Outlier Threshold Value: ',min)
print('Maximum Outlier Threshold Value: ',max)

#---------------------------------
# Index the Resonse_Time attribute
#---------------------------------
index = 'Response_time'

#--------------------------------------------------------------------------------------------------
# Plot the Resonse_Time attribute before the removal of outliers with a kernel density estimation
#--------------------------------------------------------------------------------------------------
plt.figure(figsize=(10,8))
plt.subplot(211)
plt.xlim(incident[index].min(), incident[index].max()*1.1)
plt.axvline(x=min)
plt.axvline(x=max)

# Add a kernel density estimation
ax = incident[index].plot(kind='kde')

#---------------------------------------------
# Add a Box plot to further identify outliers
#---------------------------------------------
plt.subplot(212)
plt.xlim(incident[index].min(),incident[index].max()*1.1)
sns.boxplot(x=incident[index])
plt.axvline(x=min)
plt.axvline(x=max)
```

The Minimum Outlier Threshold value is negative which does not make sense so we will set the minimum value to **0**

The Maximum Outlier Threshold value is computed as **97**

- As is evident from a visual inspection of the plots above, there are clear outliers in the Response_Time attribute

```python
#------------------------------------------------------------------------------
# We will add a new column for outliers and initialize all rows with a O value  
#------------------------------------------------------------------------------
incident['Outlier'] = 0


# update when the 
incident.loc[incident[index] < min, 'Outlier'] = 1
incident.loc[incident[index] > max, 'Outlier'] = 1

```

## Exceptional Work

To complement our Police Incident dataset we imported the following datasets to provide greater insight for our analysis.  This enabled us to better understand arrest rates against incidents as well as consider population factors that may play a critical role in effective policing of a large metropolitan city like DALLAS.

- **Police Arrests dataset**
- **Population Density dataset**

We also Geo-Mapped the data using Zip codes in Power BI to provide an enhanced visualization of crime pockets in the Dallas area. Additionally, we standardized by linking our data to the FBI National Incident-Based Reporting Systems (NIBRS) standarized Offense Codes available <a href="https://ucr.fbi.gov/nibrs/2011/resources/nibrs-offense-codes"/>  here</a>.  

Our research found that there are two categories of offenses reported in the NIBRS: Group A and Group B. It is important to determine which category an offense belongs to because the reporting
requirements for Group A offenses and Group B offenses are different. Group A offenses are used to report all incidents committed within a law enforcement agencyâ€™s jurisdiction to the
NIBRS, whereas Group B offenses are used only to report arrest data.
The following criteria were used to determine if a crime should be designated as a Group A offense:
- The seriousness or significance of the offense.
- The frequency or volume of its occurrence.
- The prevalence of the offense nationwide.
- The probability of the offense being brought to law enforcementâ€™s attention.
- The likelihood that law enforcement is the best channel for collecting data regarding the offense.
- The burden placed on law enforcement to collect data on the offense.
- The national statistical validity and usefulness of the collected data.
- The FBI UCR Programâ€™s responsibility to make crime data available not only to law enforcement but to others having a legitimate interest in it.

There are 22 Group A offense categories making up 46 Group A offenses. Each offenseâ€™s corresponding UCR Offense Code precedes its name. Immediately following the name of each offense name is an indication of whether it involves a Crime Against Person, Crime Against Property, or Crime Against Society. 

There are 11 Group B offense categories. They encompass all of the crimes that are not Group A offenses.

(Source:FBI.gov) )
