### NEW Calculated Columns added to the dataset

In order to better analyze the datset we decided to derive additional attributes as below.


```python
#-----------------------------------------------------------------
# Coding Binary variables for some columns  
#-----------------------------------------------------------------
def get_binary_code(row):
    if row == 'Yes':
        return 1
    else:
        return 0

#-----------------------------------------------------------------    
# Arrest_status variable binary coding
#-----------------------------------------------------------------
def get_status(row):
    if row['Offense_Status']=='Clear by Arrest':
        return 1
    elif row['Offense_Status']=='Clear by Exceptional Arrest':
        return 1
    else:
        return 0

#-----------------------------------------------------------------
# Converting string to datetimes
#-----------------------------------------------------------------

incident.loc[:,'Call_Received']= pd.to_datetime(incident['Call_Received_Date_Time'], 
                                                format="%m/%d/%Y %I:%M:%S %p")
incident.loc[:,'Call_Cleared']= pd.to_datetime(incident['Call_Cleared_Date_Time'], 
                                               format="%m/%d/%Y %I:%M:%S %p")
incident.loc[:,'Call_Dispatch']= pd.to_datetime(incident['Call_Dispatch_Date_Time'], 
                                                format="%m/%d/%Y %I:%M:%S %p")

#----------------------------------------------------------------------------------
# Parsing number of offenses per incident from 'Service_Number_ID' data definition
#----------------------------------------------------------------------------------
incident.loc[:,'Number_of_offense']= incident['Service_Number_ID'].str.split("-",
                                                                             expand=True).rename(columns = {2:'Number_of_offense'})


# incident.loc[:,'Year_of_offense']= incident['Service_Number_ID'].str.split("-",
#                                                                              expand=True).rename(columns = {1:'Year_of_offense'})

#----------------------------------------------------------------------------------
# Calculating response time from Call Received to Call Dispatch from original 
# columns 'Call_Received_Date_Time' and 'Call_Dispatch_Date_Time'
#----------------------------------------------------------------------------------
incident.loc[:,'Response_time'] = pd.Series(data=(incident['Call_Dispatch']-incident['Call_Received']).astype('timedelta64[m]').astype('float'),
                                          index=incident.index,name='Response_Time')


#-----------------------------------------------------------------
# Parsing Latitude and Longitude values from 'Location1 attribute 
#-----------------------------------------------------------------
incident.loc[:,'Latitude']= incident['Location1'].str.split("\n",
                                                            expand=True)[2].str.split(',',
                                                                                      expand=True)[0].str.split('(',
                                                                                                                expand=True).rename(columns = {1 : 'Latitude'}).astype(int, errors='ignore')
                                                                                                  
incident.loc[:,'Longitude']= incident['Location1'].str.split("\n",
                                                            expand=True)[2].str.split(',',
                                                                                      expand=True)[1].str.split(')',
                                                                                                                expand=True).rename(columns = {0 : 'Longitude'}).astype(int, errors='ignore')

#------------------------------------------------------------
# Incident arrest status based on "Offense_Status" attribute
#------------------------------------------------------------
incident.loc[:,'Arrest_status']= incident.apply(get_status, axis=1) 

#--------------------------------------------
# Applying Binary Coding to these attributes
#--------------------------------------------
incident.loc[:,'Hate_Crime'] = incident['Hate_Crime'].apply(get_binary_code)
incident.loc[:,'Gang_Related_Offense'] = incident['Gang_Related_Offense'].apply(get_binary_code)
incident.loc[:,'Drug_Related'] = incident['Drug_Related'].apply(get_binary_code)
```

- The following attributes were either parsed or calculated from other attributes
    + Number_of_Offense: Number of offenses per reported incident number
    + Response_Time: This is the time between receiving the call and dispatching it to officers for investigation
    + Call_Received: This is the time when the call was received by the 911 call center.
    + Call_Cleared: This is the time when the call was cleared at the 911 call center.
    + Call_Dispatch: This is the time when the call was dispatched by the 911 call center to the police officer.
    + Latitude: The latitude is parsed from the Location1 column.
    + Longitude: The longitude is parsed from the Location1 column.
    + Arrest_Status: This is binary but converted to integers Yes=1, No=0
    
    
- Following attributes were Standardized to better visualize
    + Hate_Crime

### Outlier Identification and Cleanup
- The Response time had outliers that needed to be addressed

```python
#--------------------------------------------------
# Simple Statistics of the Response time attribute
#--------------------------------------------------
incident['Response_time'].describe()
```

From the summary statistics results for the response attribute we observe the following:
- Average response time is approximately 37 minutes 
- It appears that the min is negative which does not make practical sense
- The max value also seems rather high at 13290 minutes which equates to 221 hours or 9 days
- Clearly these outliers need to be addressed before progressing with further analysis

```python
#------------------------------------------------------
# Plot IQR for Response Time - 25th and 75th quartiles
#------------------------------------------------------
q75, q25 = np.percentile(incident.Response_time.dropna(), [75 ,25])

#--------------------
# Calculate the IQR
#--------------------
iqr = q75 - q25

#-----------------------------------------------------
# Establish the limits for the outliers using the IQR
#-----------------------------------------------------
min = q25 - (iqr*1.5)
max = q75 + (iqr*1.5)

print('Minimum Outlier Threshold Value: ',min)
print('Maximum Outlier Threshold Value: ',max)

#---------------------------------
# Index the Response_Time attribute
#---------------------------------
index = 'Response_time'

#--------------------------------------------------------------------------------------------------
# Plot the Response_Time attribute before the removal of outliers with a kernel density estimation
#--------------------------------------------------------------------------------------------------
plt.figure(figsize=(10,8))
plt.subplot(211)
plt.xlim(incident[index].min(), incident[index].max()*1.1)
plt.axvline(x=min)
plt.axvline(x=max)

# Add a kernel density estimation
ax = incident[index].plot(kind='kde')

#---------------------------------------------
# Add a Box plot to further identify outliers
#---------------------------------------------
plt.subplot(212)
plt.xlim(incident[index].min(),incident[index].max()*1.1)
sns.boxplot(x=incident[index])
plt.axvline(x=min)
plt.axvline(x=max)
```

**Since the Minimum Outlier Threshold value is negative which does not make practical sense, we will set the minimum value to 0**

The Maximum Outlier Threshold value is computed as **97**

- As is evident from a visual inspection of the plots above, there are clear outliers in the Response_Time attribute

```python
#------------------------------------------------------------------------------
# We will add a new column for outliers and initialize all rows with a O value  
#------------------------------------------------------------------------------
incident['Outlier'] = 0

#---------------------------------------------------------------------------------
# Let's Update the outlier column to value=1 when the response_time is an outlier
#---------------------------------------------------------------------------------
incident.loc[incident[index] <   0, 'Outlier'] = 1
incident.loc[incident[index] > max, 'Outlier'] = 1

#--------------------------------------------
# Let's count the number of outliers
#--------------------------------------------
incident[incident.Outlier==1].Outlier.count()
```

```python
#------------------------------------------------------------------------
# We will filter out the outliers using the Outlier column where value=1
#------------------------------------------------------------------------
incident = incident[incident['Outlier']==0]

incident= incident.drop(['Outlier'],1)
```

```python
#----------------------------------------------------------------------------------------------------------------
# Re-Plot the Response_Time attribute with a kernel density estimation to validate that the outliers are removed 
#----------------------------------------------------------------------------------------------------------------
plt.figure(figsize=(10,8))
plt.subplot(211)
plt.xlim(incident[index].min(), incident[index].max()*1.1)
plt.axvline(x=min)
plt.axvline(x=max)

# Add a kernel density estimation
ax = incident[index].plot(kind='kde')

#------------------------------------------------------------
# Re-Plot the Box plot to reconfirm the removal of outliers
#------------------------------------------------------------
plt.subplot(212)
plt.xlim(incident[index].min(),incident[index].max()*1.1)
sns.boxplot(x=incident[index])
plt.axvline(x=min)
plt.axvline(x=max)
```

- As we can see, the Density Estimate plot and the box plot look much better after removal of the outliers

Our research found that there are two categories of offenses reported in the NIBRS: Group A and Group B. It is important to determine which category an offense belongs to because the reporting
requirements for Group A offenses and Group B offenses are different. Group A offenses are used to report all incidents committed within a law enforcement agency’s jurisdiction to the
NIBRS, whereas Group B offenses are used only to report arrest data.
The following criteria were used to determine if a crime should be designated as a Group A offense:
- The seriousness or significance of the offense.
- The frequency or volume of its occurrence.
- The prevalence of the offense nationwide.
- The probability of the offense being brought to law enforcement’s attention.
- The likelihood that law enforcement is the best channel for collecting data regarding the offense.
- The burden placed on law enforcement to collect data on the offense.
- The national statistical validity and usefulness of the collected data.
- The FBI UCR Program’s responsibility to make crime data available not only to law enforcement but to others having a legitimate interest in it.

There are 22 Group A offense categories making up 46 Group A offenses. Each offense’s corresponding UCR Offense Code precedes its name. Immediately following the name of each offense name is an indication of whether it involves a Crime Against Person, Crime Against Property, or Crime Against Society. 

There are 11 Group B offense categories. They encompass all of the crimes that are not Group A offenses.
